# Dataset Overview:

```python
Total Entries: The dataset contains 8,809 entries, each representing a unique movie or TV show.
Columns: There are 12 columns in the dataset:
    show_id: A unique identifier for each title.
    type: The category of the title, which is either 'Movie' or 'TV Show'.
    title: The name of the movie or TV show.
    director: The director(s) of the movie or TV show. (Contains null values for some entries, especially TV shows where this information might not be applicable.)
    cast: The list of main actors/actresses in the title. (Some entries might not have this information.)
    country: The country or countries where the movie or TV show was produced.
    date_added: The date the title was added to Netflix.
    release_year: The year the movie or TV show was originally released.
    rating: The age rating of the title.
    duration: The duration of the title, in minutes for movies and seasons for TV shows.
    listed_in: The genres the title falls under.
    description: A brief summary of the title.
```

```python
df.head()
```

```python
# removing 's' and replacing with '' from show id so that it can be converted to int dtype.
df['show_id'] = df['show_id'].str.replace('s', '')
```

```python
# changing dtype from obj to int for describe 
df['show_id']=df['show_id'].astype('int')
```

# CONTENT DISTRIBUTION

```python
# DISTRIBUTION OF WATCHING CONTENT
count_of_different_type=sns.countplot(x='type', hue='type', data=df)
for bars in count_of_different_type.containers:
    count_of_different_type.bar_label(bars)
sns.set(rc={'figure.figsize':(10,5)})
```

*Its Clearly evident more movies were produced , The count of Movies is more than that of TV Shows*

#### Ratings of Content

```python
count_of_different_type=sns.countplot(x='rating',data=df)
for bars in count_of_different_type.containers:
    count_of_different_type.bar_label(bars)
    sns.set(rc={'figure.figsize':(20,15)})
```

*MOST TV SHOWS / MOVIES were Rated as TV-MA , TV-14 and R*

```python
#Bifurcating the above barGraph.
count_of_different_type=sns.countplot(x='rating',hue='type', data=df, palette='rainbow')
for bars in count_of_different_type.containers:
    count_of_different_type.bar_label(bars)
sns.set(rc={'figure.figsize':(20,10)})
```

*The ratings of the shows and movies can be analysed easily with the above bargraph*

# TOP DIRECTORS

```python
directors_series = df['director']
director_counts = directors_series.value_counts()
top_10_directors = director_counts.head(10)
print(top_10_directors)
top_10_directors.plot(kind='bar', color='black')
plt.title('Top 10 Directors with Most Content')
plt.xlabel('Directors')
plt.ylabel('Number of Contents')
```

*Top 5 Directors with most Content*

# COUNTRY

```python
country =df['country']
count = country.value_counts()
top_5=count.head(5)
print(top_5)
```

#### Plotting pie chart for the above data for easy understanding

```python
plt.title('Top 5 countries with most content')
names=['USA' ,'INDIA', 'UK' ,'CANADA', 'Spain']

plt.pie(top_5, labels = names ,autopct='%1.2f%%')
plt.legend()
plt.show()
```

*Pie chart visualization of the most content produced by top 5 countries*
